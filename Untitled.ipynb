```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1. Generate synthetic stock price data using a Gaussian distribution
# Assume the stock prices follow a normal (Gaussian) distribution

mean_price = 100   # Mean price of the stock
std_price = 10    # Standard deviation of the stock price
n_data_points = 50 # Number of data points to generate

# Generate random normal numbers for stock prices
stock_prices = np.random.normal(mean_price, std_price, n_data_points)

plt.figure(figsize=(10, 6))
plt.hist(stock_prices, bins=20, color='blue')
plt.title('Generated Stock Prices Distribution', fontsize=14)
plt.xlabel('Stock Price (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# Save the plot to a file if you need it (requires saving the figure)
# plt.savefig('generated_stock_prices.png')
```
